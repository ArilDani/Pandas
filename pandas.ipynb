```python
print("="*5,"Operasi","="*5)
ds1 = pd.Series([2,7,11,5]) 
ds2 = pd.Series([3,4,5,6])

def OperasiData():
    operasi = {
        "penjumlahan": ds1 + ds2, #operasi penjumlahan
        "pengurangan": ds1 - ds2, #operasi pengurangan
        "perkalian": ds1 * ds2, #operasi perkalian
        "pembagian": ds1 / ds2 #operasi pembagian
    }

    for nama,hasil in operasi.items():
        print(f"{nama}:")
        print(hasil,"\n")

OperasiData()
```

```python
import re
print("="*5,"memfilter data","="*5)

print("-"*5,"Memilih Memilih Teks","-"*5)
print(df[df['nama'].str.contains("Aril")])
print(df['nama'].str.contains('Arutia'))

print("-"*5,"Abaikan Huruf Besar/Kecil","-"*5)
print(df[df['nama'].str.contains('in')])
print(df[df['nama'].str.contains('in', flags=re.IGNORECASE)])

print("-"*5,"Memilih data yang berada di list/Data frame lain","-"*5)
#menampilkan kolom februari pada file penjualan yang bernilai 130 dan 220

feb_penjualan = [130,220]
print(df[df['feb'].isin(feb_penjualan)])

print("-"*5,"Memilih angka","-"*5)

#sama dengan
print(df[df['jan']!=100])
print(df[df.jan.eq(100)])

#memilih angka tidak sama dengan
print(df[df['jan']!=100])
print(df[df.jan.ne(100)])
print(~(df['jan']==100))

#memilih angka lebih dari
print(df[df['jan']>100])
print(df[df.jan.gt(100)])

#memilih angka lebih dari atau sama dengan
print(df[df['jan']>=100])
print(df[df.jan.ge(100)])

#memilih angka kurang dari atau sama dengan
print(df[df['jan']<=100])
print(df[df.jan.le(100)])

print("-"*5,"Membuat kategori tinggi dan rendah","-"*5)
d = {'tinggi' : [340,300], 'rendah': [90,100]}
df1 = pd.DataFrame(data=d)
print(df1)

print(df[df['jan'].isin(df1['rendah'])])
print(df[df['jan'].isin(df1['tinggi'])])

#memilih lebih dari satu kondisi
##memilih diantar dua rentang nilai
print(df[(df['jan'] > 100) & (df['jan'] < 300)])
print(df[(df['nama'].str.contains('on',flags=re.IGNORECASE)) | (df['jan']>100)])
print(df.describe())
```

```python
print("="*5,"Metode apply untuk menjalankan fungsi pada setiap baris pada data series","="*5)
data = [3,3,2,4,5,10,8,13,20]
d_int = pd.Series(data)
def kecilbesar(num):
    if num < 5:
        return "kecil"
    elif num >= 5:
        return "sedang"
    else:
        return "besar"
d_int.apply(kecilbesar)
```
